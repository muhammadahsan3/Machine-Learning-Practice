```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Encoding used for categorical data

1)one-hot encoding
2)nominal encoding
3)ordinal encoding

nominal data -> each of them equal like : state names

ordinal data -> order wise : eg grades A , B , C


```python
df = pd.read_csv('customer.csv')
```

```python
df.sample(5)
```

```python
df = df.iloc[:,2:]
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:2],df.iloc[:,-1],test_size=0.2)
print(X_train.head())
```

**Ordinal Encoding**

```python
from sklearn.preprocessing import OrdinalEncoder
o_e = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
o_e.fit(X_train)
X_train = o_e.transform(X_train)
X_test = o_e.transform(X_test)
X_train
```

```python
o_e.categories_
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(y_train)
y_train = le.transform(y_train)
y_test = le.transform(y_test)
y_train
```
